## Grouping different job titles

```python
data.JobTitle.value_counts()
```

```python
def jobtitles(job):
    if 'director' in job.lower():
        job = 'Director'
    if 'clerk' in job.lower():
        job = 'Clerk'
    if 'manager' in job.lower():
        job = 'Manager'
    if 'analyst' in job.lower():
        job = 'Analyst'
    if 'assistant' in job.lower():
        job = 'Assistant'
    if 'vp' in job.lower():
        job='VP'
    else:
        job=job
    return job
```

```python
data['JobTitle']=data['JobTitle'].apply(jobtitles)
```

```python
sns.countplot(y=data.JobTitle, order = data['JobTitle'].value_counts().index)
```
